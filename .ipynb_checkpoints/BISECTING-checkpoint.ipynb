```python
def correlation(id_u,id_v):
    u = m[id_u - 1]
    v = m[id_v]
    u_movies = set([i for i in range(len(u)) if u[i] != 0])
    v_movies = set([i for i in range(len(v)) if v[i] != 0])
    common_movies = u_movies & v_movies
    uv_ratings = [[],[],[]]
    for i in common_movies:
        uv_ratings[2].append(i)
        uv_ratings[0].append(u[i])
        uv_ratings[1].append(v[i])
    uv_ratings = np.asarray(uv_ratings)
    ru = uv_ratings[0].mean();
    rv = uv_ratings[1].mean();
    num = 0
    denom = 0
    pcc = 0
    for i in range(len(uv_ratings[0])):
        num += (uv_ratings[0][i] - ru)*(uv_ratings[1][i] - rv)
        denom += math.sqrt(math.pow((uv_ratings[0][i] - ru),2)) * math.sqrt(math.pow((uv_ratings[1][i] - rv),2))
    if denom == 0:
        pcc = 1
    else:
        pcc = num / denom
    return pcc
```

```python
def forming_clusters(labels, centroids, K):
    i = 0
    clusters = []
    while i < K:
        cluster = [j for j in range(len(labels)) if labels[j] == i]
        clusters.append((cluster,centroids[i]))
         
        i += 1
    return clusters

def sse(clusters):
    sse_arr = []
    for item in clusters:
        cluster = item[0]
        centroid = item[1]
        s = 0
        for i in cluster:
            user = m[i]
            s += np.sum(np.power(np.asarray(user) - centroid, 2))
        sse_arr.append(s)
        
    return sse_arr

def kmeans(K, X):
    kmeans = KMeans(n_clusters = K, init = 'k-means++', n_init = 100,\
                    #random_state = "3", \
                    algorithm = 'full',precompute_distances=True, max_iter = 1000).fit(X)
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    return labels, centroids
```

```python
def bisecting(curr_2_clusters,sse_arr,min_K):
    k = 0
    main_cluster_arr = []
    while k < min_K:
        # следующим к разбиению на 2 кластера будет тот, у которого больше sse
        max_sse = [i for i in range(len(sse_arr)) if sse_arr[i] == max(sse_arr)][0] # в число
        cluster_to_split = curr_2_clusters[max_sse][0] # нужны элементы кластера, [1] - это уже центроиды
        # тк  cluster_to_split хранит только id пользователей, нужно достать 
        # сами значения
        users = [m[i] for i in cluster_to_split]

        new_model = kmeans(K, users)
        curr_2_clusters = forming_clusters(new_model[0], new_model[1], K)
        sse_arr = sse(curr_2_clusters)

        main_cluster_arr.append(curr_2_clusters)
        main_sse.append(sse_arr)

        k += 1
    
    return main_cluster_arr
```
